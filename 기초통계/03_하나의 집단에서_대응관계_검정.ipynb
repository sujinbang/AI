```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### paired Samples t-test
#하나의 집단에서 보통 전/후 의 관계를 확인할 때 사용
#사용되는 함수: ttest_rel(x,y)
# x = before, y=after
#-검증 :p-value > 0.05 (전후 상관없음 = 차이없음)
#      :p-value <=0.05 (전후 상관있음 = 차이있음)
```

### exe-1) 야식을 먹기전과 먹은후의 몸무게 차이 검정
##### -귀무가설 : 야식을 먹기전과 먹은 후의 몸무게에 차이가 없다.
##### -대립가설 : 야식을 먹기전과 먹은 후의 몸무게에 차이가 있다.

```python
#야식 복용 전 몸무게 랜덤 생성
beforeWeights = np.random.randint(low=40, high =90, size=20)
print('beforeWeights = ', beforeWeights)
print()

#야식 복용 후 몸무게 랜덤 생성
temp_list = []
for weight in beforeWeights :
    # -0.99-0.02 ~ 0.99+0.02
    temp_list.append(weight * np.random.normal(0.99,0.02))
print('temp_list = ', temp_list)

afterWeights = np.array(temp_list)
print('afterWeights = ',afterWeights)

# 전/후 검증하기
result = stats.ttest_rel(beforeWeights, afterWeights)

print('검정 통계량 =',  result)
print('p-value  = %.3f, p-value = %.3f ' %(result))

#(해석 -검증결과)
# 1.p-value = 0.023 < 0.05 이므로, 귀무가설 기각
# 2. 대립가설 채택 : 전/후 데이터의 차이는 있다.
#                  : 야식을 먹기전과 먹은 후는 몸무게에 차이가 발생한다.
```

### exe-2) 특강 전/후 시험점수의 차이 검정
##### -귀무가설 : 특강 전/후 시험점수의 차이가 없다.
##### -대립가설 : 특강 전/후 시험점수의 차이가 있다.

```python
import numpy as np
import scipy.stats as stats
from seaborn.distributions import distplot
```

```python
# 데이터 생성하기
#랜덤값 시드(seed) 임의로 설정
np.random.seed(1234)

#특강전 데이터 : 랜덤하게 80을 기준으로 +-10, 100개 생성
x1 = np.random.normal(80, 10, 100)
print('특강전 x1 = ', x1)

#특강 후 데이터 : 랜덤하게 77을 기준으로 +-10, 100개 생성
x2 = np.random.normal(77, 10, 100)
print('특강전 x2 = ', x2)
```

```python
# 정규성 검정 (stats.shapiro())

resultSh = stats.shapiro(x1)
print('정규성 검정 결과 = ', resultSh)
print('검정 통계량 = %.3f, p-value = %.3f' %(resultSh))

# (해석)
# - p-value = 0.047 > 0.05 이므로, 정규성을 만족한다

resultSh = stats.shapiro(x2)
print('정규성 검정 결과 = ', resultSh)
print('검정 통계량 = %.3f, p-value = %.3f' %(resultSh))

# (해석)
# - p-value = 0.747 > 0.05 이므로, 정규성을 만족한다
```

```python
#정규성 시각화(한글처리 포함)

# 한글 처리
plt.rc('font', family = 'Malgun Gothic')

sns.distplot(x1, fit=stats.norm)
sns.distplot(x2, fit=stats.norm)
plt.show()
```

```python
### 등분산성 검증
# - 귀무가설 : 두 집단간 분산이 동일하다.(p-value > 0.05)
# - 대립가설 : 두 집단간 분산이 동일하지 않다.

# 등분산성 검증은 3가지 방법으로 진행됩니다.
#1. levene(레빈) 방법
#2. fligner(플리그너) 방법
#3. bartlett(바틀렛) 방법

# 1.
sLevene = stats.levene(x1, x2)
print('정규성 검정 결과 =',  sLevene)
print('t-검정 통계량 = %.3f, p-value = %.3f ' %(sLevene))
print(sLevene)
#-p -value 는 0.560 >0.05이므로 귀무가설에 채택함
# 귀무가설 채택: 두 집단간 등분산성이 만족을 한다..

# 2.
sFligner = stats.fligner(x1, x2)
print('정규성 검정 결과 =',  sFligner)
print('t-검정 통계량 = %.3f, p-value = %.3f ' %(sFligner))
print(sFligner)
#-p -value 는 0.471 >0.05이므로 귀무가설에 채택함
# 귀무가설 채택: 두 집단간 등분산성이 만족을 한다..

# 3.
sBartlett = stats.bartlett(x1, x2)
print('정규성 검정 결과 =',  sBartlett)
print('t-검정 통계량 = %.3f, p-value = %.3f ' %(sBartlett))
print(sFligner)
#-p -value 는 0.863 >0.05이므로 귀무가설에 채택함
# 귀무가설 채택: 두 집단간 등분산성이 만족을 한다..

# 등분산성 검증은 두 집단간에 관계가 있는지를 확인하는 것

# 한 집단에서 전후관계를 검정하는 것이기때문에 등분산성 검증을 원래는 하지않음
```

```python
# 전/후 검증하기
result = stats.ttest_rel(x1, x2)

print('검정 결과 =',  result)
print('검정 통계량 = %.3f, p-value = %.3f ' %(result))

#(해석 -검증결과)
# 1.p-value = 0.005 < 0.05 이므로, 귀무가설 기각
# 2. 대립가설 채택 : 전/후 데이터의 차이는 있다.
#                  : 특강 전과 후는 시험점수 차이가 발생한다.
```

### exe-3) 환자 9명의 복부 수술전/후 몸무게 변화에 대한 차이 검정
##### - 귀무가설 : 복부 수술 전/후 몸무게 변화가 없다
##### - 대립가설 : 복부 수술 전/후 몸무게 변화가 있다

```python
#랜덤값 시드(seed) 임의로 설정
np.random.seed(136514)

# 데이터 생성
t1 = []
x1 = np.random.normal(80, 20, 9)
for i in x1:
    t1.append(round(i, 1))
print('수술전 x1 = ', t1)

t2 = []
x2 = np.random.normal(70, 15, 9)
for i in x2:
    t2.append(round(i, 1))
print('수술후 x2 = ', t2)
```

```python
# 정규성 검정 (stats.shapiro())

resultSh = stats.shapiro(t1)
print('정규성 검정 결과 = ', resultSh)
print('검정 통계량 = %.3f, p-value = %.3f' %(resultSh))

# (해석)
# - p-value = 0.939 > 0.05 이므로, 정규성을 만족한다

resultSh = stats.shapiro(t2)
print('정규성 검정 결과 = ', resultSh)
print('검정 통계량 = %.3f, p-value = %.3f' %(resultSh))

# (해석)
# - p-value = 0.182 > 0.05 이므로, 정규성을 만족한다
```

```python
#정규성 시각화(한글처리 포함)

# 한글 처리
plt.rc('font', family = 'Malgun Gothic')

sns.distplot(t1, fit=stats.norm)
sns.distplot(t2, fit=stats.norm)
plt.show()
```

```python
# 전/후 검증하기
result = stats.ttest_rel(t1, t2)

print('검정 결과 =',  result)
print('검정 통계량 = %.3f, p-value = %.3f ' %(result))

#(해석 -검증결과)
# 1.p-value = 0.228 > 0.05 이므로, 귀무가설 채택
# 2. 귀무가설 채택 : 전/후 데이터의 차이는 없다.
#                  : 수술 전과 후는 시험점수 차이가 없다.
```
